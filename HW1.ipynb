```python
import numpy as np
import matplotlib.pyplot as plt

ts = [0,1,2]

x = np.linspace(-5,5,100)

def f(x,t):
    return 1 / np.cosh(x-t)
for t in ts:
    y = f(x,t)
    plt.plot(x,y, label=f"$t = {t}$")
plt.xticks(np.arange(-5,5.1,1))
plt.legend()
plt.grid()
plt.title("$u(x,t)=sech^2(x-t) \\quad x \\in [-5,5]$")
```

```python
js = np.arange(1,11.1,1)

x = np.linspace(-2,6,100)

def f2(x):
    sums = 0
    for j in js:
        sums = sums + (((-1)**j)/np.sqrt(j))*np.cos((j*np.pi*x)/2)
    return sums
y = f2(x)
plt.plot(x,y)
plt.xticks(np.arange(-2,6.1,1))
plt.grid()
plt.title("$\\sum_{j=1}^{11} \\frac{(-1)^j}{\\sqrt{j}} \\cos \\left( \\frac{j \\pi x}{2} \\right) \\text{ over } x \\in [-2, 6]$")
plt.show()
```
