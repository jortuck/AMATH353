```python
# Question 2 Part A
import numpy as np
import matplotlib.pyplot as plt

ts = [0,1,2,3]

x = np.linspace(-5,5,100)

def f(x,t):
    return 0.5*(np.exp(-(x-2*t)**2)+np.exp(-(x+2*t)**2) + (np.cos(x-2*t) +np.cos(x+2*t))/4)
for t in ts:
    y = f(x,t)
    plt.plot(x,y, label=f"$t = {t}$")
plt.xticks(np.arange(-5,5.1,1))
plt.legend()
plt.grid()
plt.title("$u(x,t)=\\frac{e^{-(x-2t)^2} + e^{-(x+2t)^2}}{2} + \\frac{\\cos(x-2t) + \\cos(x+2t)}{4} \\quad x \\in [-5,5]$")
```

```python
# Question 2 Part B
import numpy as np
import matplotlib.pyplot as plt

ts = [0,1,2,3]

x = np.linspace(-5,5,500)

def f(x,t):
    return 0.5*((np.cos((x-2*t)**2) +np.cos((x+2*t)**2)) + (np.arctan(x+2*t) - np.arctan(x-2*t))/2)
for t in ts:
    y = f(x,t)
    plt.plot(x,y, label=f"$t = {t}$")
plt.xticks(np.arange(-5,5.1,1))
plt.legend()
plt.grid()
plt.title(r"$u(x,t) = \frac{\cos((x-2t)^2) + \cos((x+2t)^2)}{2} + \frac{\arctan(x+2t) - \arctan(x-2t)}{4} \quad x \in [-5,5]$")
```

```python
# Question 3 Part B
import numpy as np
import matplotlib.pyplot as plt

# Updated time steps
ts = [0, 1, 2, 3, 4, 5]

# Updated domain to x >= 0 (using [0, 10] based on earlier setup)
# Increased points to 1000 because x^50 creates very sharp features
x = np.linspace(0, 10, 1000)

def f(x, t):
    # Helper for the initial displacement function f(x)
    def f_init(v):
        return (v**2) * (1 / (1 + v**50) + 1 / (10 + (v - 3)**20))

    # Using np.where to handle the piecewise condition (c=1, g(x)=0)
    # np.where(condition, true_result, false_result)
    return np.where(x >= t,
                    0.5 * (f_init(x - t) + f_init(x + t)),
                    0.5 * (f_init(t - x) + f_init(x + t)))

for t in ts:
    y = f(x, t)
    plt.plot(x, y, label=f"$t = {t}$")

# Updated ticks to match the new domain
plt.xticks(np.arange(0, 10.1, 1))
plt.legend()
plt.grid()


plt.show() # Added plt.show() so the graph renders if running as a script
```
