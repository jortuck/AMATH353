```python
import numpy as np
import matplotlib.pyplot as plt

ts = [0, 1, 2, 3, 4]
x = np.linspace(0, 3, 100)
c = 1/4

def f(x, t):
    term1 = 3/8
    term2 = 0.5 * np.cos(4 * np.pi / 3 * x) * np.cos(4 * c * np.pi / 3 * t)
    term3 = 0.125 * np.cos(8 * np.pi / 3 * x) * np.cos(8 * c * np.pi / 3 * t)
    return term1 + term2 + term3

for t in ts:
    y = f(x, t)
    plt.plot(x, y, label=f"$t = {t}$")

plt.xticks(np.arange(0, 3.1, 0.5))
plt.legend()
plt.grid()
plt.title(r"$u(x,t)=\frac{3}{8} + \frac{1}{2}\cos\left(\frac{4\pi}{3}x\right)\cos\left(\frac{\pi}{3}t\right) + \frac{1}{8}\cos\left(\frac{8\pi}{3}x\right)\cos\left(\frac{2\pi}{3}t\right) \quad x \in [0,3]$")

plt.show()
```
