```python
# Question 2 Part A
import numpy as np
import matplotlib.pyplot as plt

ts = [0,1,2,3]

x = np.linspace(-5,5,100)

def f(x,t):
    return 0.5*(np.exp(-(x-2*t)**2)+np.exp(-(x+2*t)**2) + (np.cos(x-2*t) +np.cos(x+2*t))/4)
for t in ts:
    y = f(x,t)
    plt.plot(x,y, label=f"$t = {t}$")
plt.xticks(np.arange(-5,5.1,1))
plt.legend()
plt.grid()
plt.title("$u(x,t)=e^{-(x-2t)^2} + e^{-(x+2t)^2} + \\frac{\\cos(x-2t) + \\cos(x+2t)}{4} \\quad x \\in [-5,5]$")
```

```python
# Question 2 Part B
import numpy as np
import matplotlib.pyplot as plt

ts = [0,1,2,3]

x = np.linspace(-5,5,500)

def f(x,t):
    return 0.5*((np.cos((x-2*t)**2) +np.cos((x+2*t)**2)) + (np.arctan(x+2*t) - np.arctan(x-2*t))/2)
for t in ts:
    y = f(x,t)
    plt.plot(x,y, label=f"$t = {t}$")
plt.xticks(np.arange(-5,5.1,1))
plt.legend()
plt.grid()
plt.title("$u(x,t)=e^{-(x-2t)^2} + e^{-(x+2t)^2} + \\frac{\\cos(x-2t) + \\cos(x+2t)}{4} \\quad x \\in [-5,5]$")
```
